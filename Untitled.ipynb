# in this line im going to calculate bmi of people

# Formula for BMI : BMI=WeightÂ (kg)/(HeightÂ (m))^2

# formula for converting cm to m = height(m) = height/100



```python
# if height given in cm
data_f["Height_m"] = data_f["Height"] / 100  

# BMI calculation
data_f["BMI"] = data_f["Weight"] / (data_f["Height_m"] ** 2)
data_f
```

# making seperate column for bmi category  

# column name is fitness

```python
def fitness_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"

data_f["Fitness"] = data_f["BMI"].apply(fitness_category)
data_f
```

```python
data_f[["Height", "Weight", "BMI", "Fitness"]]
```

# in this scatter plot im showing fitness level according to height and weight 

```python
mlt.figure(figsize=(10,6))
sns.scatterplot(
    x="Height", 
    y="Weight", 
    hue="Fitness",        # Fitness category ke hisaab se color
    data=data_f[["Height", "Weight", "BMI", "Fitness"]], 
    palette="Set2",       # colors scheme
    s=100,                # point size
    edgecolor="black"
)

mlt.title("Height vs Weight with Fitness Category")
mlt.xlabel("Height (cm)")
mlt.ylabel("Weight (kg)")
mlt.legend(title="Fitness Level")
mlt.show()
```

 # Pie plot for those who have a diabetic and blood pressure problem or not

```python


# --- Diabetes Pie ---
diabetes_counts = data_f['Diabetes'].value_counts()
labels_dia = ['No Diabetes', 'Has Diabetes']

mlt.figure(figsize=(6,6))
mlt.pie(diabetes_counts, labels=labels_dia, autopct='%1.2f%%',
        startangle=90, shadow=True, explode=(0, 0.1))
mlt.title("Diabetes Distribution")
mlt.show()


# --- Blood Pressure Pie ---
bp_counts = data_f['BloodPressureProblems'].value_counts()
labels_bp = ['No BP Problem', 'Has BP Problem']

mlt.figure(figsize=(6,6))
mlt.pie(bp_counts, labels=labels_bp, autopct='%1.1f%%',
        startangle=90, shadow=True, explode=(0, 0.1))
mlt.title("Blood Pressure Problems Distribution")
mlt.show()
```
